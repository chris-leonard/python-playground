# Stochastic Processes

Notes on stochastic processes from a probabilistic perspective.

## Measure Theory

**Definition:** A *$\sigma$-algebra* $\mathcal{F}$ on a set $\Omega$ is a non-empty subset of the powerset $\mathcal{P}(\Omega)$ of $\Omega$ that is closed under complements, countable unions, and countable intersections. The pair $\left(\Omega, \mathcal{F}\right)$ is called a *measureable space* and the elements of $\mathcal{F}$ are called *measurable sets*.

**Examples:**
1. If $\Omega$ is any non-empty set then $\lbrace \emptyset, \Omega\rbrace$ is a $\sigma$-algebra on $\Omega$.
2. If $\Omega$ is any non-empty set then $\mathcal{P}\left(\Omega\right)$ is a $\sigma$-algebra on $\Omega$.
3. If $\Omega=\lbrace a, b, c, d\rbrace$ then $\lbrace \emptyset, \lbrace a, b\rbrace, \lbrace c, d\rbrace, \Omega$ is a $\sigma$-algebra on $\Omega$.

**Definition:** If $\mathcal{F}$ is any subset of $\mathcal{P}(\Omega)$ then there is a unique smallest $\sigma$-algebra $\sigma\left(\mathcal{F}\right)$ on $\Omega$ containing $\mathcal{F}$; the intersection of all $\sigma$-algebras on $\Omega$ containing $\mathcal{F}$. This is called the *$\sigma$-algebra generated by $\mathcal{F}$*.

**Example:** Suppose that $\Omega$ is a topological space. The *Borel $\sigma$-algebra* $\mathcal{B}$ on $\Omega$ is the $\sigma$-algebra generated by all open (or all closed) sets in $\Omega$. The elements of $\mathcal{B}$ are called *Borel sets*. If $\Omega=\mathbb{R}$ with the usual topology then $\mathcal{R}$ is generated by all open intervals $(a, b)$, $a\in\mathbb{R}$ (and by all half-open intervals $(a, b]$, and by all intervals $(a, \infty)$).

**Definition:** Suppose that $\Omega$ and $S$ are non-empty sets, $\Sigma$ is a $\sigma$-algebra on $S$, and $f:\Omega\to S$ is a function. Let $\sigma(f)$ denote the set of pre-images of elements of $\Sigma$:

\begin{equation}
    \sigma(f) = \lbrace\, f^{-1}(E) \,\mid\, E\in\Sigma \,\rbrace.
\end{equation}

Then $\sigma(f)$ is a $\sigma$-algebra on $\Omega$ called the *$\sigma$-algebra generate
d by $f$*.


**Definition:** Suppose that $\left(\Omega, \mathcal{F}\right)$ and $\left(S, \Sigma\right)$ are measurable spaces and $f:\Omega\to S$. We call $f$ a *measurable function*, or more precisely a *$\left(\mathcal{F}, \Sigma\right)$-measurable function*, if $\sigma(f)\subseteq \mathcal{F}$. That is, if
$f^{-1}(E)\in\mathcal{F}$ for all $E\in\Sigma$.
