# Stochastic Processes

Notes on stochastic processes from a probabilistic perspective.

## Measure Theory

**Definition:** A *$\sigma$-algebra* $\mathcal{F}$ on a set $\Omega$ is a non-empty subset of the powerset $\mathcal{P}(\Omega)$ of $\Omega$ that is closed under complements, countable unions, and countable intersections. The pair $\left(\Omega, \mathcal{F}\right)$ is called a *measureable space* and the elements of $\mathcal{F}$ are called *measurable sets*.

**Examples:**
1. If $\Omega$ is any non-empty set then $\lbrace \emptyset, \Omega\rbrace$ is a $\sigma$-algebra on $\Omega$.
2. If $\Omega$ is any non-empty set then $\mathcal{P}\left(\Omega\right)$ is a $\sigma$-algebra on $\Omega$.
3. If $\Omega=\lbrace a, b, c, d\rbrace$ then $\lbrace \emptyset, \lbrace a, b\rbrace, \lbrace c, d\rbrace, \Omega$ is a $\sigma$-algebra on $\Omega$.

**Definition:** If $\mathcal{F}$ is any subset of $\mathcal{P}(\Omega)$ then there is a unique smallest $\sigma$-algebra $\sigma\left(\mathcal{F}\right)$ on $\Omega$ containing $\mathcal{F}$; the intersection of all $\sigma$-algebras on $\Omega$ containing $\mathcal{F}$. This is called the *$\sigma$-algebra generated by $\mathcal{F}$*.

**Example:** Suppose that $\Omega$ is a topological space. The *Borel $\sigma$-algebra* $\mathcal{B}\left(\Omega\right)$ on $\Omega$ is the $\sigma$-algebra generated by all open (or all closed) sets in $\Omega$. The elements of $\mathcal{B}$ are called *Borel sets*. If $\Omega=\mathbb{R}$ with the usual topology then $\mathcal{B}\left(\mathbb{R}\right)$ is generated by any out of:
- all open intervals $(a, b)$, $a\in\mathbb{R}$;
- all half-open intervals $(a, b]$, $a, b\in\mathbb{R}$;
- all intervals $(a, \infty)$, $a\in\mathbb{R}$;
- all intervals $(-\infty, b]$, $b\in\mathbb{R}$.

**Definition:** Suppose that $\Omega$ and $S$ are non-empty sets, $\Sigma$ is a $\sigma$-algebra on $S$, and $f:\Omega\to S$ is a function. Let $\sigma(f)$ denote the set of pre-images of elements of $\Sigma$:

\begin{equation}
    \sigma(f) = \lbrace\, f^{-1}(E) \,\mid\, E\in\Sigma \,\rbrace.
\end{equation}

Then $\sigma(f)$ is a $\sigma$-algebra on $\Omega$ called the *$\sigma$-algebra generated by $f$*.


**Definition:** Suppose that $\left(\Omega, \mathcal{F}\right)$ and $\left(S, \Sigma\right)$ are measurable spaces and $f:\Omega\to S$. We call $f$ a *measurable function*, or more precisely a *$\left(\mathcal{F}, \Sigma\right)$-measurable function*, if $\sigma(f)\subseteq \mathcal{F}$. That is, if
$f^{-1}(E)\in\mathcal{F}$ for all $E\in\Sigma$.

## Probability

### Probability Spaces

**Definition:** A *probability space* is a triple $\left(\Omega, \mathcal{F}, P\right)$ where:
- $\Omega$ is a non-empty set called the *sample space*;
- $\mathcal{F}$ is a $\sigma$-algebra on $\Omega$ called the *event space*, whose elements are *events*;
- $P$ is a *probability measure* on $\left(\Omega, \mathcal{F}\right)$; a function $\mathcal{F}\to[0, 1]$ such that:
    - $P(\Omega)=1$;
    - $P$ is *$\sigma$-additive*: if $E_n\in\mathcal{F}$ ($n\in\mathbb{N}$) is a countable collection of pair-wise disjoint events then
    
\begin{equation}
    P\left( \bigcup_{n\in\mathbb{N}} E_i\right) = \sum_{n\in\mathbb{N}} P\left(E_n\right).
\end{equation}

**Example:** Consider a coin flip where the coin has a probability $p\in[0, 1]$ of coming up heads. This can be represented by the probability space $\left(\Omega, \mathcal{F}, P\right)$ with $\Omega=\lbrace H, T\rbrace$, $\mathcal{F}=\mathcal{\Omega}=\lbrace \emptyset, \lbrace H\rbrace, \lbrace T\rbrace, \lbrace H, T\rbrace\rbrace$, and $P$ the unique probability measure with $P(H)=p$ and $P(T)=1-p$.

**Intuition:** We should think of $\Omega$ as the outcomes that can happen and $\mathcal{F}$ as the information that's available, or what outcomes can be distinguished from each other. Often $\mathcal{F}=\Omega$, but it can be useful to consider smaller $\sigma$-algebras that represent the outcomes that can be determined with partial information. 

**Example:** Consider a sequence of two flips of a coin with probability $p$ of coming up heads. This situation is described by the probability space with $\Omega=\lbrace HH, HT, TH, TT\rbrace$, $\mathcal{F}=\mathcal{P}\left(\Omega\right)$, and $P$ the unique probability measure determined by

\begin{equation}
    P(HH)=p^2, \qquad P(HT)=P(TH)=1-p, \qquad P(TT)=(1-p)^2.
\end{equation}

But suppose we wish to describe the observed information after the first flip. The outcomes $HH$ and $HT$ can't be distinguished with this limited information, so shouldn't be considered distinct events. Instead, define a new event space $\mathcal{F}_1 = \lbrace \emptyset, \lbrace HH, HT\rbrace, \lbrace TH, TT\rbrace, \Omega \rbrace$. The probability space $\left(\Omega, \mathcal{F}_1, P\big|_{\mathcal{F}_1}\right)$ represents this information that can be determined at this point.

### Random Variables

**Definition:** Let $\left(\Omega, \mathcal{F}, P\right)$ be a probability space and $\left( S, \Sigma\right)$ be a measurable space. A *random variable*, or more precisely a *$\left( S, \Sigma\right)$-valued random variable* is a $\left(\mathcal{F}, \Sigma\right)$-measurable function $X:\Omega\to S$.

**Notation:** If $A\in\Sigma$ is a measureable set then we define

\begin{equation}
    P\left( X\in A\right) \mathrel{\mathop:}= P\left( X^{-1}(A)\right).
\end{equation}

If $A=\{a\}$ then we will write $P\left(X=a\right)$ in place of $P\left( X\in A\right)$. Finally, if $A=(-\infty, b]$ (or its intersection with $S$) we write $P\left( X\leq b\right)$ in place of $P\left( X\in A\right)$. This extends to other inequalities in the obvious way.

**Examples:**
1. Let $S$ be a countable set and $\Sigma=\mathcal{P}(S)$, or equivalently let $\Sigma$ be the Borel $\sigma$-algebra $\mathcal{B}(S)$ of $S$ where $S$ is endowed with the discrete topology (all sets are both open and closed). Since $\Sigma$ is generated (as a $\sigma$-algebra) by all singletons, a function $X:\Omega\to S$ is a random variable if and only if $P\left(X=a\right)$ is well-defined for all $a\in S$. If so, we call $X$ a *discrete random variable*. Particularly important examples are $S=\mathbb{N}$ and $S=\mathbb{Z}$.
2. Let $S=\mathbb{R}$ with its usual topology and set $\Sigma=\mathcal{B}\left(\mathbb{R}\right)$. A function $X:\Omega\to \mathbb{R}$ is a random variable if and only $P(X\leq b)$ is well-defined for all $b\in\mathbb{R}$ (or $P(a<X<b)$ is well-defined for all $a, b\in\mathbb{R}$, etc.). If so, we call $X$ a *real-valued random variable*, or simply a *real random variable*.

Note that since singletons are closed in $\mathcal{B}\left(\mathbb{R}\right)$, if $S$ is a countable subset of $\mathbb{R}$ then the two examples coincide. In this case we might call $X$ a *discrete real-valued random variable*.

**Convention:** With real random variables the sample and event spaces are almost incidental and we often describe real-valued random variables solely in terms of the probability measure $P$, without any reference to $\Omega$ and $\mathcal{F}$.

**Examples:**
1. Consider a single coin flip with a probability $p$ of coming up heads and define a probability space $\left(\Omega, \mathcal{F}, P\right)$ as before. Define a discrete real random variable $X$ by $X(H)=1$ and $X(T)=0$. Then $P\left(X=1\right)=p$ and $P\left(X=1\right)=1-p$.
2. Suppose you flip the coin above 10 times in a row. Let $\Omega$ denote the set of 10 letter words in the alphabet $\lbrace H, T\rbrace$, $\mathcal{F}=\mathcal{P}\left(\Omega\right)$, and let $P$ determined by $P(H)=p$ and $P(T)=1-p$ in the obvious way. Suppose you are betting on the outcome of the coin flips: you win £1 for every head and lose £2 for every tail. Let $X$ be the discrete real random variable that represents you winnings at the end of the game. So, for example, $P(X=-20)=(1-p)^{10}$.
